```python
def check_input(x):
  if type(x) == str:
    return True
  else:
    return False


def get_substring(x):
  dlina = 1
  podstroka = x[0]
  maks = dlina
  maks_stroka = podstroka
  for i in range(1,len(x)-1):
    if x[i] != x[i-1]:
      dlina = dlina + 1
      podstroka = podstroka + x[i]
      if dlina > maks:
        maks = dlina
        maks_stroka = podstroka
    else:
      dlina = 1
      podstroka = x[i]
  return maks_stroka


def main(x):
  stroka = ''
  maks = 0
  ind = 0
  for i in range(len(x)):
    if check_input(x[i]) == True:
      if len(x[i]) > maks:
        maks = len(x[i])
        stroka = x[i]
        ind = i
  if stroka == '':
    return -1, ''
  else:
    return ind, get_substring(stroka)
```
