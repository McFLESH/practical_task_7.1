```python
def check_input(x):
  if type(x) == str:
    return True
  else:
    return False


def get_substring(x):
  dlina = 1
  stroka = x[0]
  maks = dlina
  maks_stroka = stroka
  for i in range(1,len(x)-1):
    if x[i] != x[i-1]:
      dlina = dlina + 1
      stroka = stroka + x[i]
      if dlina > maks:
        maks = dlina
        maks_stroka = stroka
    else:
      dlina = 1
      stroka = x[i]
  return maks_stroka


def main(x):
  stroki = []
  for i in range(len(x)):
    if check_input(x[i]) == True:
      stroki.append(x[i])
  return stroki
print(main([[0, 4, 8], 'abca', 7, 'aaaaa']))
```
