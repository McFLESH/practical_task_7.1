```python
def get_substring(x):
  dlina = 1
  stroka = x[0]
  maks = dlina
  maks_stroka = stroka
  for i in range(1,len(x)-1):
    if x[i] != x[i-1]:
      dlina = dlina + 1
      stroka = stroka + x[i]
      if dlina > maks:
        maks = dlina
        maks_stroka = stroka
    else:
      dlina = 1
      stroka = x[i]
  return maks_stroka
print(get_substring('aaaaffwvvbw'))
```
